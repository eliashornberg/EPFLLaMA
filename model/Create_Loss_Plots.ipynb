```python
import json
import matplotlib.pyplot as plt
import utils

# Load JSON data from a file
data = utils.read_json("imgs/SFT/train_Sft.json")

# Extract steps and loss values
steps = [item[1] for item in data]
loss = [item[2] for item in data]

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(steps, loss, marker='o')
plt.title('Training Loss Over Steps', size=16)
plt.xlabel('Step', size=14)
plt.ylabel('Loss', size=14)
plt.grid(True)
plt.savefig('imgs/SFT/train_loss.png', dpi = 300,  bbox_inches='tight')
```
